# Python basic Course

In this basic python course, we will dicuss the basics of data type, 
data structure, flow control, and  functions 


```python
# Assignment
# ==========

# Strings
data = 'hello world'
print(data[0])
print(len(data))
print(data)
```

```python
# Numbers
value = 123.1
print(value)
value = 10
print(value)
```

```python

# Boolean
a = True
b = False
print(a, b)
```

```python
# Multiple Assignment
a, b, c = 1, 2, 3
print(a, b, c)
```

```python
#Flow Control
# ============

# If-Then-Else

value = 99
if value >= 99:
	print('That is fast')
elif value > 200:
	print('That is too fast')
else:
	print('That that is safe')
```

```python
# For-Loop
for i in range(10):
	print(i)
```

```python
# While-Loop
i = 0
while i < 10:
	print(i)
	i += 1
```

```python
# Data Structures
# ===============

# Tuple (cannot change)
a = (1, 2, 3)
print(a)
```

```python
# Lists
mylist = [1, 2, 3]
print("Zeroth Value: %d") % mylist[0]
mylist.append(4)
print("List Length: %d") % len(mylist)
for value in mylist:
	print(value)
```

```python
# Dictionaries

mydict = {'a': 1, 'b': 2, 'c': 3}
print("A value: %d") % mydict['a']
mydict['a'] = 11
print("A value: %d") % mydict['a']
print("Keys: %s") % mydict.keys()
print("Values: %s") % mydict.values()
for key in mydict.keys():
	print(mydict[key])
```

```python
# Functions
# ===========

# Sum function
def mysum(x, y):
	return x + y

# Test sum function
mysum(1, 3)
```
